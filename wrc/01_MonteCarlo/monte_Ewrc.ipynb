```python
import urllib.request
from bs4 import BeautifulSoup as soup
import requests
import re
import pandas as pd
link = 'https://www.ewrc-results.com/results/77787-rallye-automobile-monte-carlo-2023/?s='
start=375528
#monte_23 = pd.DataFrame(columns=['Pos.', 'Pilote / Co-pilote', 'Voiture', '#', 'Temps', 'Écart'])
monte_23 = []
```

```python
my_url11 = link + str(start)
req = urllib.request.Request(my_url11, headers={'User-Agent': 'Mozilla/5.0'})
uClient11 = urllib.request.urlopen(req)
page_html11 = uClient11.read()
uClient11.close()
data = pd.read_html(page_html11)[0]
data.head()
```

```python
data.set_axis(['SS_Position','No', 'Crew','Gr', 'SS_Time','diff','Speed'], axis='columns', inplace=True)
#data.set_axis(['SS_Position','No', 'Crew','Gr', 'SS_Time','diff','Speed'], axis= 'columns', copy=False)
for col in data.columns:
    print(col)
#data.to_csv('test1.csv',index=False)
```

```python
for ss in range(1,7):
    ss_a = str(ss+start)
    my_url11 = link + ss_a
    req = urllib.request.Request(my_url11, headers={'User-Agent': 'Mozilla/5.0'})
    uClient11 = urllib.request.urlopen(req)
    page_html11 = uClient11.read()
    uClient11.close()
    data = pd.read_html(page_html11)[0]
    data['ss']=ss
    monte_23.append(data)
```

```python
len(monte_23)
```

```python
monte2023_stages = pd.concat(monte_23, axis=0)
```

```python
monte2023_stages.set_axis(['SS_Position','No', 'Crew','Gr', 'SS_Time','diff','Speed','SS'], axis= 'columns', inplace=True)
monte2023_stages.to_csv('test2.csv',index=False)
monte2023_stages.info()
```
