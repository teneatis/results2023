```python
import urllib.request
from bs4 import BeautifulSoup as soup
import requests
import re
import pandas as pd
link = 'http://rallyliveresults.com/wrc/live-result/epreuves-speciales-special-stages/?stage='
```

```python
for ss in range (1,19):
    ss_a = str(ss)
    my_url11 = link+ss_a
    req = urllib.request.Request(my_url11, headers={'User-Agent': 'Mozilla/5.0'})
    uClient11 = urllib.request.urlopen(req)
    page_html11 = uClient11.read()
    uClient11.close()
    
    
    page_soup2 = soup(page_html11, "html.parser")
    table = page_soup2.find('table')
    table = str(table)
    table = re.sub(r'<img.*?>','', table)
    table = re.sub(r'\n','', table)
    table = soup(table, "html.parser")
    #print(my_url11)
    headers = []
    for i in table.find_all('th'):
        title = i.text
        headers.append(title)

    monte_2023 = pd.DataFrame(columns=headers)
 
    for j in table.find_all('tr')[1:]:
        row_data = j.find_all('td')
        row = [i.text for i in row_data]
        length = len(monte_2023)
        monte_2023.loc[length]=row
    
    monte_2023['ss']=ss
    table='-'
    req = ''
    uClient11 = ''
    page_html11 = ''
    print(monte_2023.head())
    monte_2023.to_csv('monte23_SS'+ss_a+'.csv',index=False)
    monte_2023.drop(monte_2023.index, inplace=True)
    #print(monte_2023)
```
