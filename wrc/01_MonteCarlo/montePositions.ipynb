```python
import urllib.request
from bs4 import BeautifulSoup as soup
import requests
import re
import pandas as pd
link = 'http://rallyliveresults.com/wrc/live-result/epreuves-speciales-special-stages/?stage='
#monte_23 = pd.DataFrame(columns=['Pos.', 'Pilote / Co-pilote', 'Voiture', '#', 'Temps', 'Écart'])
monte_23 = []
```

```python
for ss in range(1,19):
    ss_a = str(ss)
    my_url11 = link + ss_a
    req = urllib.request.Request(my_url11, headers={'User-Agent': 'Mozilla/5.0'})
    uClient11 = urllib.request.urlopen(req)
    page_html11 = uClient11.read()
    uClient11.close()
    data = pd.read_html(page_html11)[0]
    data['ss']=ss
    monte_23.append(data)
```

```python
monte2023_stages = pd.concat(monte_23, axis=0)
```

```python
monte2023_temp=monte2023_stages.drop(['Écart','Temps'], axis=1)
monte2023_view = monte2023_temp.set_index(['#','Pilote / Co-pilote','Voiture','ss'], drop=True).unstack('ss')
```

```python
monte2023_temp2 = monte2023_temp[monte2023_temp['#']<30]
monte2023_view2 = monte2023_temp2.set_index(['#','Pilote / Co-pilote','Voiture','ss'], drop=True).unstack('ss')
```

```python
monte2023_stages.to_csv('monte2023_AllStages.csv', index=False)
```
