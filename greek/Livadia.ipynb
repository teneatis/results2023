```python
import pandas as pd
data = pd.read_csv('Livadia2023_st__01.csv')
data = data.fillna('-')
data
```

```python
#data['No'] = data['No'].replace('#','',inplace=True)
data['No'] = data['No'].str[1:]
data
```

```python
#dataview = data[['No', 'ss_time','ss']].copy()
dataview = data.drop(['Diff','Speed','Pos.'], axis=1)
dataview = dataview.fillna('-')
dataview['No'] = dataview['No'].astype(str).astype(int)
data_view2 = dataview.set_index(['No', 'Crew','Gr/Cl','ss'], drop=True).unstack('ss')
data_view2 = data_view2.fillna('-')

dataview2 = data_view2.sort_values(by=['No'])
#dataview2.to_csv('Livadia2023.csv', encoding='utf-8', index=False)
dataview2.to_csv('Livadia2023_a.csv', encoding='utf-8')
dataview2

```
