```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import glob

f = 'greek_rallies_2023_stages_times.csv'
```

```python
data = pd.read_csv(f)
data2 = data[data['Gr/Cl'].str.startswith('C')]
data2
```

```python
data2['GroupC'] = data2['Gr/Cl'].str[:2]
data2 
```

```python
sns.boxplot(x="GroupC", y="time_SS_sec", data=(data2))

plt.show()
```

```python
data2
```

```python
figsize=(140, 140)
exp_plot = sns.catplot(data=data2, kind="box", y="time_SS_sec", x='SS.', hue='GroupC')
exp_plot.figure.savefig('output_figure.png',dpi=600)
```

```python

sns.violinplot(x="GroupC", y="time_SS_sec", data=(data2[data2['Race']=='Potamies']), native_scale=True,hue='ss')

plt.show()
```

```python

sns.violinplot(x="GroupC", y="time_SS_sec", data=(data2), native_scale=False)

plt.show()
```
