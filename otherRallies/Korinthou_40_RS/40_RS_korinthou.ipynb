```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import glob

ss_n = 1
data =[]
files = []
for f in glob.glob('*_split.csv'):
    files.append(f)
    temp_df = pd.read_csv(f)
    temp_df['ss'] = ss_n
    data.append(temp_df)
    print(f'Δημιουργήθηκε dataframe για το {f} με {temp_df.shape}')
    ss_n+=1

total = pd.concat(data, axis=0)
stage_times = pd.read_csv('Korinthou_Stages_Times_Sec.csv')
stage_times['No'] = stage_times['No'].str.replace('#','').astype(str).astype(int)
stage_times['Gr/Cl'] = stage_times['Gr/Cl'].str[:2]
stage_times
```

```python
total.dtypes
```

```python
split_times = pd.merge(total, stage_times, on=['No','ss'])
split_times
```

```python
split_t = split_times.iloc[:, 0:40:2]
split_t['Crew'] = split_times['Crew']
split_t['ss'] = split_times['ss']
split_t = split_t.drop(['Race', 'surface', 'cou'], axis=1)
split_t
```

```python
split_s = split_times.iloc[:,  1:40:2]
split_s['Crew'] = split_times['Crew']
split_s['No'] = split_times['No']
split_s = split_s.drop(['date', 'competition'], axis=1)
split_s
```

```python
filter_col=[]
filter_col = [col for col in split_s if col.startswith('CP')]
s=1
k=l=m=""
for col in filter_col:
    k=str(s)+"_sec"
    l=str(s)+"_cen"
    m=str(s)+"_time_Sec"
    split_s[k]=split_s[col].str[:2].astype(str).astype(float)
    split_s[l]=split_s[col].str[3:].astype(str).astype(float)
    split_s[m]=split_s[k]*60 + split_s[l]
    split_s = split_s.drop([col, k, l], axis=1)
    #split_s = split_s.drop([col], axis=1)
    s+=1
    k=l=m=""
split_s
split_s.to_csv('split_s.csv', index=False)
```
